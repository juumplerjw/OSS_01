```python
size = 5

stack = [None for _ in range(size)]

top = -1

print(stack)
```

```python
def isstackfull():
    global size, stack, top
    if (top >= size-1):
        return True
    else:
        return False
```

```python
def push(data):
    if (isstackfull()):
        print("full")
        return
    top += 1
    stack[top] = data
```

```python
def isstackempty():
    global size, stack, top
    if (top == -1):
        return True
    else:
        return False
```

```python
def pop():
    global size, stack, top
    if (isstackempty()):
        print("empty")
        return None
    data = stack[top]
    stack[top] = None
    top -= 1
    return None
```

```python
def peek():
    global size, stack, top
    if(isstackempty()):
        print("empty")
        return None
    return stack[top]
```
